# Remove & Rename Columns

```python
columns_to_drop = ['Q13 - Ethnicity', 'Q8 - If you were to look for a new job today, what would be the most important thing to you?', 'Q7 - How difficult was it for you to break into Data?', 'Q6 - How Happy are you in your Current Position with the following? (Upward Mobility)', 'Email', 'Referrer', 'Time Taken (America/New_York)', 'Browser', 'OS', 'City', 'Country']
df = df.drop(columns=columns_to_drop, axis = 1)
```

```python
new_column_names = {'Q1 - Which Title Best Fits your Current Role?':'Job Title', 'Q2 - Did you switch careers into Data?':'Career Switch to Data', 'Q3 - Current Yearly Salary (in USD)':'Annual Salary (USD)', 'Q4 - What Industry do you work in?':'Working Industry', 'Q5 - Favorite Programming Language':'Favorite Programming Language', 'Q6 - How Happy are you in your Current Position with the following? (Salary)':'Happy with salary', 'Q6 - How Happy are you in your Current Position with the following? (Work/Life Balance)':'Happy with work-life balance', 'Q6 - How Happy are you in your Current Position with the following? (Coworkers)':'Happy with coworkers', 'Q6 - How Happy are you in your Current Position with the following? (Management)':'Happy with management', 'Q6 - How Happy are you in your Current Position with the following? (Learning New Things)':'Happy with learning opportunities', 'Q9 - Male/Female?':'Gender', 'Q10 - Current Age':'Age', 'Q11 - Which Country do you live in?':'Country', 'Q12 - Highest Level of Education':'Level of Education'}
df = df.rename(columns=new_column_names)
```

# Change Data Type

```python
df['Date Taken (America/New_York)'] = pd.to_datetime(df['Date Taken (America/New_York)'])
```

```python
df
```

```python
# drop duplicates in case of any

df.drop_duplicates()
```

# Perform Cleaning if others are specified

```python
print(df['Job Title'].unique())
```

```python
df['Job Title'] = df['Job Title'].apply(lambda x: x.split(':', 1)[-1].strip() if ':' in x else x)

df['Working Industry'] = df['Working Industry'].apply(lambda y: y.split(':', 1)[-1].strip() if ':' in y else y)

df['Favorite Programming Language'] = df['Favorite Programming Language'].apply(lambda z: z.split(':', 1)[-1].strip() if ':' in z else z)

df['Country'] = df['Country'].apply(lambda s: s.split(':', 1)[-1].strip() if ':' in s else s)
```

# Modify Salary Column: Create Average Salary Column

```python
print(df['Annual Salary (USD)'].unique())
```

```python
df[['Min Annual Salary', 'Max Annual Salary']] = df['Annual Salary (USD)'].str.split('-', expand=True)

df['Min Annual Salary'] = df['Min Annual Salary'].str.replace('k', '000').str.replace('+', '').astype(float)
df['Max Annual Salary'] = df['Max Annual Salary'].str.replace('k', '000').str.replace('+', '').astype(float)

df['Average Annual Salary'] = (df['Min Annual Salary'] + df['Max Annual Salary']) / 2
```

```python
print(df[df['Annual Salary (USD)'] == '225k+'])
```

```python
df['Max Annual Salary'].fillna(225000.0, inplace=True)

df['Min Annual Salary'].replace(225000.0, 0.0, inplace=True)

df['Average Annual Salary'].fillna(112500.0, inplace=True)
```

# Loop through "favorite programming language" column (answers given were a bit messy)

```python
programming_languages = ['Python', 'R', 'SQL', 'JavaScript', 'C/C++', 'C#', 'Excel', 'Java', 'SAS', 'VBA', 'MATLAB', 'Php']

for index, entry in df['Favorite Programming Language'].items():
    for lang in programming_languages:
        if lang.lower() in entry.lower():
            df.loc[index, 'Favorite Programming Language'] = lang
            break
```

```python
# assign null entries to be just empty

df['Level of Education'] = df['Level of Education'].fillna('')
```
